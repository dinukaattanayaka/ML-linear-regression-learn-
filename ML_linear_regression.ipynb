```python
import pandas as pd

dataset=pd.read_csv('/content/drive/MyDrive/machine learning/DAY 6 (Regression)/weight-vs-height.csv').values
print(dataset)
```

```python
data=dataset[:,0].reshape(-1,1)
target=dataset[:,1]
```

```python
from matplotlib import pyplot as plt

plt.scatter(data,target)
plt.xlabel('Feature (Weights(Kg)-x)')
plt.ylabel('Label (Heights(cm)-y)')
```

```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(train_data,train_target)
```

```python
print('m:',model.coef_)
print('c:',model.intercept_)
```

```python
test_weight=[[85]]
yb=model.predict(test_weight)
print('predicted height:',yb)
```
